```python
url = "https://api.coingecko.com/api/v3/coins/markets"
params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1,
    "sparkline": "true",
    "price_change_percentage": "24h,7d"
}

response = requests.get(url, params=params)
print("Status code:", response.status_code)
data = response.json()
print(type(data), len(data) if isinstance(data, list) else data)
df = pd.json_normalize(data)
print(df.columns.tolist())
df.head()
```

```python
df = pd.json_normalize(data)

df = df[[
    "id", "symbol", "current_price", "market_cap",
    "total_volume", "price_change_percentage_24h",
    "price_change_percentage_7d_in_currency",
    "sparkline_in_7d.price"
]]

df.columns = ["coin", "symbol", "price_usd", "market_cap",
              "volume_24h", "change_24h", "change_7d", "sparkline"]

df
```

```python
fig = go.Figure()
for i, row in df.iterrows():
    fig.add_trace(go.Scatter(
        y=row["sparkline"],
        mode="lines", name=row["coin"]
    ))
fig.update_layout(title="7-Day Price Trend — Top 10 Coins", template="plotly_dark",
                   xaxis_title="Hours (last 7 days)", yaxis_title="Price (USD)")
fig.show()
```

```python

sparkline_dict = {row["coin"]: row["sparkline"] for _, row in df.iterrows()}


lengths = {k: len(v) for k, v in sparkline_dict.items()}
print(lengths)


min_len = min(lengths.values())
sparkline_dict = {k: v[:min_len] for k, v in sparkline_dict.items()}

price_matrix = pd.DataFrame(sparkline_dict)
corr = price_matrix.corr()

fig = px.imshow(corr, text_auto=".2f", color_continuous_scale="RdBu_r",
                 title="7-Day Price Correlation Between Top 10 Coins")
fig.show()
```
